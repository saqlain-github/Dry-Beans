Network Intrusion Detection Systems (NIDSs) have become an important component in network security infrastructure. Currently, many NIDSs are rule-based systems whose performances highly depend on their rule sets. Unfortunately, due to the huge volume of network traffic, coding the rules by security experts becomes difficult and time-consuming. Since data mining techniques can build intrusion detection models adaptively, data mining-based NIDSs have significant advantages over rule-based NIDSs .

As per or anaysis, the Network Intrusion Detection is a classic calssification problem.

Classification is the problem of identifying to which of a set of categories (subpopulations), a new observation belongs, on the basis of a training set of data containing observations and whose categories membership is known.

Classification is of two types:

1. Binary Classification: When we have to categorize given data into 2 distinct classes. Example – On the basis of given health conditions of a person, we have to determine whether the person has a certain disease or not.

2. Multiclass Classification: The number of classes is more than 2. For Example – On the basis of data about different species of flowers, we have to determine which specie does our observation belongs to.

#### CLASS IMBALANCE CHECK -

Perform class imblance check on pre-processed data file based on target column. 

Most machine learning algorithms work best when the number of samples in each class are about equal. This is because most algorithms are designed to maximize accuracy and reduce errors.
However, if the data set in imbalance then In such cases, you get a pretty high accuracy just by predicting the majority class, but you fail to capture the minority class, which is most often the point of creating the model in the first place.

```python
df_base_file['class'].value_counts()/df_base_file.shape[0]
```

We see an imbalance here. ~57% of data belongs t anomaly class where as only ~43% data belongs to normal class. This implies that more than 50% of dataset that we have recieved is anomalies.
We will use few steps to handle this class imbalance in classification problem:-
1. Use stratification to select train-test-validation datasets. Use Stratified K-Fold split.
2. In case we use XGBOOST/CatBoost or LightGBM, we can use scale_pos_weight. If we are using sklearn, we can use class imbalance dictionary.
3. The evaluation metric is ROC-AUC but in case F1/Precision/Recall was to be used, I would have set a probabilistic threshold on true class probabilities (probability of minority class) and trained the model.

#### Creation of test and validation sets -

We have sufficient data available. We create 3 subsets out of the data given to us-

Train data - This will be 90% of the data given to us on which the model will be trained. While performing Stratified K fold cross validation, a certain percentage of it will be used as a validation dataset.

Test data - This will be the remaining 10% of the data where the final metrics will be reported.

```python
X_train, X_test_val, y_train, y_test_val = train_test_split(base_x, cat_Ytest, test_size=0.1, random_state=2,stratify=cat_Ytest)
```

```python
print(f'Shape of train data will be      {X_train.shape}')
print(f'Shape of validation data will be {X_test_val.shape}')
```

```python
train      = X_train.values
train_y    = y_train.values
test       = X_test_val.values
y_test_val = y_test_val.values
```

## Model 1:- Baseline Logistic model
We fit a Baseline logistic model without any parameter tuning

```python
oof_pred               = np.zeros((len(train), ))
y_pred                 = np.zeros((len(test), ))

y_pred_final           = np.zeros((len(test), 2))

n_splits               = 10
error                  = []
kf                     = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=2)

for fold, (tr_ind, val_ind) in enumerate(kf.split(train, train_y)):   
    
    X_train, X_val     = train[tr_ind], train[val_ind]
    y_train, y_val     = train_y[tr_ind], train_y[val_ind]


    
    model      = LogisticRegression(random_state=2,
                                    class_weight = 'balanced',   ### handles class imbalance here 
                                    max_iter=1000,
                                    )
    _          = model.fit(X_train,y_train)
    
    val_pred   = model.predict_proba(X_val)[:,1]
    
    roc        = roc_auc_score(y_val,val_pred)
    recall     = recall_score(y_val, np.where(val_pred>0.4,1,0))
    precision  = precision_score(y_val, np.where(val_pred>0.4,1,0))
    accuracy   = accuracy_score(y_val, np.where(val_pred>0.4,1,0))
    f_score    = f1_score(y_val, np.where(val_pred>0.4,1,0))
    
    
    print(f'Fold {fold+1}')
    print(f'Validation ROC AUC score = {roc}, Recall = {recall}, Precision = {precision}, Accuracy = {accuracy}, F-Score = {f_score}')
    
    oof_pred[val_ind]  = val_pred
    y_pred    += model.predict_proba(test)[:,1] / (n_splits)
print('\n')
print('OOF(Cross-Validation) ROC_AUC :- ',(roc_auc_score(train_y,oof_pred)))
print('\n')
print('ROC AUC on test data selected :- ',(roc_auc_score(y_test_val,y_pred)))
```

## Observations from the baseline logistic model:- 
#### Positive points :- 
1. Performs decently well accross all folds and the ROC score is not changing drastically accross all folds.
2. The cross validation score and the score on test dataset are pretty close. So our validation data is representative of the test data and we are probably not overfitting 
3. Class imbalance taken care of 

#### Negative points:- 
1. Scores can be improved. (Scores can be considered still low)
2. No parameter tuning done as of now. (many features are redundant and perhaps, L1 regularization will have a role to play.

#### To do list :- 
Let's convert the above segment of code to a function so that we can use it iteratively and then parameter tune it 

```python
def custom_cross_val(model_type,args,verbose=False):
  
  oof_pred               = np.zeros((len(train), ))
  y_pred                 = np.zeros((len(test), ))

  y_pred_final           = np.zeros((len(test), 2))

  n_splits               = 10
  error                  = []
  kf                     = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=2)

  for fold, (tr_ind, val_ind) in enumerate(kf.split(train, train_y)):   
      
      X_train, X_val     = train[tr_ind], train[val_ind]
      y_train, y_val     = train_y[tr_ind], train_y[val_ind]


      if model_type =='logistic':
        model      = LogisticRegression(**args)

      _          = model.fit(X_train,y_train)
      
      val_pred   = model.predict_proba(X_val)[:,1]
      
      roc        = roc_auc_score(y_val,val_pred)
      recall     = recall_score(y_val, np.where(val_pred>0.4,1,0))
      precision  = precision_score(y_val, np.where(val_pred>0.4,1,0))
      accuracy  = accuracy_score(y_val, np.where(val_pred>0.4,1,0))
      f_score  = f1_score(y_val, np.where(val_pred>0.4,1,0))
      
      if verbose:
        print(f'Fold {fold+1}')
        print(f'Validation ROC AUC score = {roc}, Recall = {recall}, Precision = {precision}, Accuracy = {accuracy}, F-Score = {f_score}')
      
      oof_pred[val_ind]  = val_pred
      y_pred    += model.predict_proba(test)[:,1] / (n_splits)
  return roc_auc_score(train_y,oof_pred),roc_auc_score(y_test_val,y_pred)
```

```python
penalty_vals = ['l1','l2']
C            = [0.5,1,1.5,2,3]
for penalty in penalty_vals:
  for c in C:
    args = {'random_state':2,'penalty':penalty,'solver':'liblinear','class_weight':'balanced','max_iter':1000,'C':c}
    cv,test_score = custom_cross_val('logistic',args)
    print(f'CV score : {cv}, Test score: {test_score}, Corresponding arguments {args},  ')
```

### Best Logistic Model:- 

We can clearly see that the best Logistic model is the one with the parameters - <br>
{'random_state': 2, 'penalty': 'l1', 'solver': 'liblinear', 'class_weight': 'balanced', 'max_iter': 1000, 'C': 2}

Let's incorporate extra tree classifier as well in the custom function (which includes logistic regression) and see how good we are able to perform

```python
def custom_cross_val(model_type,args,verbose=False):
  
  oof_pred               = np.zeros((len(train), ))
  y_pred                 = np.zeros((len(test), ))

  

  n_splits               = 10
  error                  = []
  kf                     = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=2)

  for fold, (tr_ind, val_ind) in enumerate(kf.split(train, train_y)):   
      
      X_train, X_val     = train[tr_ind], train[val_ind]
      y_train, y_val     = train_y[tr_ind], train_y[val_ind]


      if model_type =='logistic':
        model      = LogisticRegression(**args)
      if model_type=='extra_trees':
        model      = ExtraTreesClassifier(**args)

      _          = model.fit(X_train,y_train)
      
      val_pred   = model.predict_proba(X_val)[:,1]
      
      roc        = roc_auc_score(y_val,val_pred)
      recall     = recall_score(y_val, np.where(val_pred>0.4,1,0))
      precision  = precision_score(y_val, np.where(val_pred>0.4,1,0))
      accuracy   = accuracy_score(y_val, np.where(val_pred>0.4,1,0))
      f_score    = f1_score(y_val, np.where(val_pred>0.4,1,0))
      
      if verbose:
        print(f'Fold {fold+1}')
        print(f'Validation ROC AUC score = {roc}, Recall = {recall}, Precision = {precision}, Accuracy = {accuracy}, F-Score = {f_score}')
    
      
      oof_pred[val_ind]  = val_pred
      y_pred    += model.predict_proba(test)[:,1] / (n_splits)
  return roc_auc_score(train_y,oof_pred),roc_auc_score(y_test_val,y_pred)
```

#### Let's perform a Grid Search to find optimal parameter set.

Note :- If we look at the combinations to test, we can see clearly we can have 160 distinct parameter sets to test. <br>
It will be a long code to process. But we will let it run and in the next step, move with Random Search and then a combination of Grid and Random Search

```python
estimators   = [50,100,150,200]
depth        = [4,5,6,8,None]
max_features = ['auto','sqrt','log2',None]
bootstrap    = [True,False]
max_val      =  0
best_args    = None
for est in estimators:
  for dep in depth:
    for max_feat in max_features:
      for bootst in bootstrap:
        args = {'random_state':2,
                'class_weight':'balanced',
                'n_estimators':est,
                'max_depth':dep,
                'bootstrap':bootst,
                'max_features':max_feat
                }
        cv,test_score = custom_cross_val('extra_trees',args)
        if test_score>max_val:
          max_val = test_score
          best_args = args
        print(f'CV score : {cv}, Test score: {test_score}, Corresponding arguments {args},  ')
```

```python
print('The best arguments attained are :- ')
print(best_args)
print('Maximum test score attained was :- ',max_val)
```

### Observation from Grid Search.
We see that the best results are attained when estimator is much less and depth reached the maximum. We need to see if we increase the depth a bit more, we see a better test score was attained than before. 

```python
args = {'random_state': 2, 'class_weight': 'balanced', 'n_estimators': 100,
        'max_depth': 10, 'bootstrap': True, 'max_features': None,
        'n_jobs':-1}
cv,test_score = custom_cross_val('extra_trees',args,verbose=True)
print(f'CV score : {cv} and test score : {test_score}')
```

### Custom Random Search
From the information gained from above, we will now perform Custom Random search on the given set of parameters for 10 iterations and see the performance

```python
estimators   = [100,120,110]
depth        = [8,9,10,None]
bootstrap    = [True,False]
max_val      =  0
best_args    = None

for iterations in range(10):
  est      = random.choice(estimators)
  dep      = random.choice(depth)
  bootst   = random.choice(bootstrap)
  args = {'random_state':2,
          'class_weight':'balanced',
          'n_estimators':est,
          'max_depth':dep,
          'bootstrap':bootst,
          'max_features':None,
          'n_jobs':-1
          }
  cv,test_score = custom_cross_val('extra_trees',args)

  if test_score>max_val:
    max_val = test_score
    best_args = args
  print(f'iteration {iterations+1},CV score : {cv}, Test score: {test_score}, Corresponding arguments {args},  ')
```

### Conclusion.
We clearly can see that we were marginally able to improve the performance as compared to the one attained via grid search and the test roc auc score of around 0.9996 is the maximum we can get from the ExtraTrees Classifier from the given set of features we have

#### Comparison of both parameter tuning strategies.
* While Grid search performs a search on all possible parameter sets available, it takes a longer time to execute and hence may not be effective when we have a huge data and a complex model (from here on model complexity will only increase). Complexity is defined in terms of number of parameters to tune.
* Random search is effective and fast and the ideal procedure must be to use random search for a finite set of iterations and then apply Grid search on the neighbourhood model to reduce the parameter search space for Grid search.

Learning from our experience, we will now perform Random Search on the parameters and then select the best one

```python
def custom_cross_val(model_type,args,verbose=False):
  
  oof_pred               = np.zeros((len(train), ))
  y_pred                 = np.zeros((len(test), ))

  
  n_splits               = 10
  error                  = []
  kf                     = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=2)

  for fold, (tr_ind, val_ind) in enumerate(kf.split(train, train_y)):   
      
      X_train, X_val     = train[tr_ind], train[val_ind]
      y_train, y_val     = train_y[tr_ind], train_y[val_ind]


      if model_type =='logistic':
        model      = LogisticRegression(**args)
      if model_type=='extra_trees':
        model      = ExtraTreesClassifier(**args)
      if model_type =='xgb':
        model      = XGBClassifier(**args)

      _          = model.fit(X_train,y_train)
      
      val_pred   = model.predict_proba(X_val)[:,1]
      
      roc        = roc_auc_score(y_val,val_pred)
      
      if verbose:
        print(f'Fold {fold+1}, Validation ROC AUC score {roc}')
      
      oof_pred[val_ind]  = val_pred
      y_pred    += model.predict_proba(test)[:,1] / (n_splits)
  return roc_auc_score(train_y,oof_pred),roc_auc_score(y_test_val,y_pred)
```

## Random Search on XGBoostClassifier parameters
We perform random search on the parameters of XGBoost

```python
estimators   = [50,100,200,300]
depth        = [4,6,8]
subsample    = [0.7,0.8,0.9,1]
reg_lambda   = [1,5,10]
lr           = [0.07,0.09,0.10,0.05]

max_val      =  0
best_args    = None

for iterations in range(20):
  est      = random.choice(estimators)
  dep      = random.choice(depth)
  learn    = random.choice(lr)
  sub      = random.choice(subsample)
  regl     = random.choice(reg_lambda)

  args = {'random_state':2,
          'scale_pos_weight':spw,
          'n_estimators':est,
          'max_depth':dep,
          'learning_rate':learn,
          'subsample':sub,
          'reg_lambda':regl,
          'use_label_encoder':False,
          'eval_metric':'logloss'
          }
  cv,test_score = custom_cross_val('xgb',args)

  if test_score>max_val:
    max_val = test_score
    best_args = args
  print(f'iteration {iterations+1},CV score : {cv}, Test score: {test_score}, Corresponding arguments {args},  ')
```

# Conclusions and Recommendations - 

We performed classification of given data using three different classifiers - 

1. Logistic Linear Regression: This algorithm is a machine learning algorithm by examining the relationship between dependent variables and independent variables, just like in linear regression, and the biggest difference from classical linear regression is data classification (Rogel-Salazar, 2018; Sarkar, 2019). It is a special form of regression, and although the name of the algorithm is referred to as regression, it is used for classification rather than estimation. For this reason, logistic regression method is also known as binary classification (Ozdemir, 2016). Logistic regression has a low variance due to its simple operation structure and is therefore less prone to over fitting.

2. Extra Trees Classifier: In this model, a series of decision trees are created from various subsets of the training data set (Rogel-Salazar, 2018). It is a classification in which averages are used to check the accuracy of the prediction by estimating the model that fits the random decision tree.

3. XG Boost Classifier: As with classical gradient enhancement models, it is a decision tree-based community learning algorithm (Rufaida et al., 2020). It was revealed in 2016 and it's considered new. The main difference from other algorithms lies in its scalability, which enables fast learning through parallel and distributed computing and provides efficient memory usage. It is free from over fitting and bias.

We have seen that Extra Trees Classifier gives better performance than Logistic Linear Regression and XG Boost Classifier gives best performance out of all three of these.
